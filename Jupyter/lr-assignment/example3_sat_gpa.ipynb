```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Change to data path on your computer
data = pd.read_csv('SAT_GPA.csv')

# Show the description of data
data.describe()

# Set to training data (x, y)
y_data = data['GPA']
x_data = data['SAT']

# Remind that we need to put component x_0 = 1 to x
plt.scatter(x_data, y_data)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
# Extracting the first 60 data points
X = x_data.values[:60]
y = y_data.values[:60]

# Calculate mean of SAT and GPA
X_mean = np.mean(X)
y_mean = np.mean(y)
print('Mean of SAT:', X_mean)
print('Mean of GPA:', y_mean)
```

```python
# Method: OLS
# Calculate the sum of squared deviations from the mean
SS_xx = np.sum((X - X_mean) ** 2)
SS_xy = np.sum((X - X_mean) * (y - y_mean))
print('SXX:', SS_xx)
print('SXY:', SS_xy)
```

```python
# Calculate slope and intersect of y = b + ax
slope = SS_xy / SS_xx
intercept = y_mean - slope * X_mean
print('slope:', slope)
print('intercept:', intersect)
```

```python
# Plot the regression line
regression_line = slope * X + intercept
plt.scatter(x_data, y_data)
plt.plot(X, regression_line, color = 'red')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
# --- R-squared, SSE, and MSE calculations ---

# Predicted values
y_pred = slope * X + intercept

# SSE (Sum of Squared Errors)
SSE = np.sum((y - y_pred) ** 2)

# SST (Total Sum of Squares)
SST = np.sum((y - y_mean) ** 2)

# R-squared
R_squared = 1 - (SSE / SST)

# MSE (Mean Squared Error)
MSE = SSE / len(y)

# Print the results
print(f'SSE (Sum of Squared Errors): {SSE}')
print(f'MSE (Mean Squared Error): {MSE}')
print(f'R-squared: {R_squared}')
```
